```python
from ultralytics import YOLO

# Load a model YOLO v 11
modelv11 = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)
```

```python
# Train the model with GPUs
resultsV11 = modelv11.train(data="/kaggle/input/blood-cell-detection-datatset/data.yaml", epochs=100, imgsz=640, device=[0])
```

```python
metricsV11 = modelv11.val()  # Ini akan mengembalikan dict dengan mAP, precision, recall, dll
# print(resultsV11.metrics)
print(metricsV11)
```
